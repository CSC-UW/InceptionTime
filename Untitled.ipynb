```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.load('/mnt/nfs/proj/in-vitro/Leonardo/cf_data/CF082.npz')
x_probe,y_probe = data['x'], data['y']
_, ulidx = np.unique(y_probe, return_index=True, axis=0)
print(len(ulidx))
```
